```python
 import pandas as pd
 import numpy as np
 from sklearn.model_selection import train_test_split
 from sklearn.ensemble import RandomForestClassifier

```

```python
df=pd.read_csv("/content/dataset.csv")
```

```python
df
```

```python


df.fillna(0)
```

```python
l1=df["Symptom_1"].tolist()
ab=list(df.columns)
symptoms=[]
for i in range(1,len(ab)):
    l1=df[ab[i]].tolist()
    symptoms=symptoms+l1
  
abs=set(symptoms)
len(abs)
l3=[x for x in range(1,134)]
dic_b=dict(zip(abs,l3))
dic_b["null"]=0
dic_a=dict(zip(l3,abs))
dic_b

```

```python
for i in range(1,len(ab)):
    df[ab[i]].replace(dic_b.keys(),dic_b.values(),inplace=True)
    df[ab[i]].replace(125,0,inplace=True)
```

```python
df
```

```python
diseases=set(df["Disease"].tolist())
print(len(diseases))
list3=[x for x in range(1,42)]
dic_c=dict(zip(list3,diseases))
df["Disease"].replace(dic_c.values(),dic_c.keys(),inplace=True)
dic_c
```

```python
dic_b

```

```python
Y=df['Disease']
X=df.drop('Disease',axis=1)
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,random_state=104,test_size=0.25,shuffle=True)
clf4=RandomForestClassifier(n_estimators=100)
clf4.fit(X_train,Y_train)
```

```python
def getList(listA):
  predl=[]
  array=list(np.zeros(17))
  for i in range(len(listA)):
    array[i]=dic_b[listA[i]]
  predl.append(array)
  return predl
```

```python
pred_res=clf4.predict(getList(["itching"," skin_rash"," nodal_skin_eruptions"," dischromic _patches"]))
```
